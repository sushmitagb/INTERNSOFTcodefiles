```python
def askint():
    while True:
        try:
            val = int(input("Please enter an integer: "))
        except:
            print("Looks like you did not enter an integer!")
            continue
        else:
            print('Yep thats an integer!')
            break
        finally:
            print("Finally, I executed!")
        print(val) 
```

```python
askint()
```

 DATABASE CONNECTIVITY AND OPERATION USING PYTHON

```python
#!/usr/bin/python

import sqlite3


db = sqlite3.connect("my_database1.db")


db.execute("drop table if exists grades1")


db.execute("create table grades1(id int, name text, score int)")


db.execute("insert into grades1(id, name, score) values(116, 'naira',95 )")
db.execute("insert into grades1(id, name, score) values(118, 'ayra',93 )")
db.execute("insert into grades1(id, name, score) values(119, 'myra', 99 )")
db.execute("insert into grades1(id, name, score) values(120, 'kaira', 90 )")
db.execute("insert into grades1(id, name, score) values(121, 'zara',88 )")
```

```python
db.commit()
```

```python
results = db.execute("select * from grades1 order by id")
for row in results:
    print(row)
print("-" * 60 )
```

```python
results = db.execute("select * from grades1 where name = 'Gary' ")
for row in results:
    print(row)
print("-" * 60 )
```

```python
results = db.execute("select * from grades1 where score >= 90 ")
for row in results:
    print(row)
print("-" * 60 )
```

```python
results = db.execute("select name, score from grades1 order by score desc ")
for row in results:
    print(row)
print("-" * 60 )
```

```python
results = db.execute("select name, score from grades1 order by score")
for row in results:
    print(row)
print("-" * 60 )
```
