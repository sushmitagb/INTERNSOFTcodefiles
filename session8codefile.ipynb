```python
normals = pd.Series(np.random.normal(size=50))
print(normals)
normals.plot()
```

```python
normals.cumsum().plot(grid=True)
```

```python
variables = pd.DataFrame({'normal': np.random.normal(size=20), 
                       'gamma': np.random.gamma(2, size=20), 
                       'poisson': np.random.poisson(size=20)})
print(variables)

variables.cumsum(0).plot()
```

```python
variables.cumsum(0).plot(subplots=True)
```

```python
variables.cumsum(0).plot(secondary_y='normal')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(5, 30))
for i,var in enumerate(['normal','gamma','poisson']):
    variables[var].cumsum(0).plot(ax=axes[i], title=var)
axes[0].set_ylabel('cumulative sum')
```

```python
fig = plt.figure()
ax = plt.axes()

s = np.linspace(0, 50, 200)
print(s)
ax.plot(s, np.sin(s));
```

```python
plt.plot(s, np.sin(s))
```

```python
plt.plot(s, np.sin(s - 0), color='green')        
plt.plot(s, np.sin(s - 1), color='y')           
plt.plot(s, np.sin(s - 2), color='1.55')       
```

```python
plt.plot(s, np.sin(s - 2), linestyle='-')        
plt.plot(s, np.sin(s - 3), linestyle=':')           
plt.plot(s, np.sin(s - 4), linestyle='-.')  
```

```python
plt.plot(s, np.sin(s - 0))

plt.xlim(-5, 30)
plt.ylim(-4, 4);
```

```python
plt.plot(s, np.cos(s))
plt.title("Cosine Curve")
plt.xlabel("x")
plt.ylabel("cos(x)")
```

```python
rng = np.random.RandomState(9)
x = rng.randn(400)
y = rng.randn(400)
colors = rng.rand(400)
sizes = 2000 * rng.rand(400)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.4,
            cmap='Reds')
plt.colorbar()  # show color scale
```

```python
x = np.linspace(0, 10, 100)
dy = 0.1
y = np.sin(x) + dy * np.random.randn(100)

plt.errorbar(x, y, yerr=dy, fmt='.b');
```
